Getting real pricing data yFinance 

```python
NVDA_df = yf.download('NVDA', 
                  start='2017-01-01', 
                  end='2024-12-31', 
                  progress=False)
```

```python
X = NVDA_df['Close']
```

```python
X.index.array
```

```python

plt.plot(X.index.array,X.values)
plt.ylabel('Close')
plt.legend(['Nvidia'])
```

Percentage change - getting rid of the first value due to nan val

```python
R = X.pct_change()[1:]
```

```python
plt.hist(R,bins=20)
plt.xlabel("PReturns")
plt.ylabel("Frequancy")
plt.legend("NVIDIA Returns")
```

```python
np.mean(R)
```

```python
np.std(R)
```

```python
plt.hist(np.random.normal(np.mean(R), np.std(R), 10000), bins=20)
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(['Normally Distributed Returns']);
```
